```python
nanocomposites=pd.read_csv('NANOCOMPOSITES_DATASET_here.csv')
pd.set_option('display.max_rows',10) #vedo tutte le righe mettendo None (lo posso fare anche con le colonne, ma qui è inutile)
nanocomposites
```

```python
nanocomposites.columns
```

```python
nanocomposites_properties=nanocomposites[['Polymer Matrix', 'Filler', 'Filler Concentration (wt%)',
       'Elastic Modulus (MPa)', 'Tensile Strength (MPa)', 'Tensile Strain (%)', 'Elastic Modulus matrix (MPa)']]
```

```python
nanocomposites_properties.isna().sum()
```

```python
nanocomposites_properties.describe() #describe è un metodo che crea un riassunto statistico delle colonne numeriche di questo dataframe
```

```python
nanocomposites_properties.dtypes #dtypes attributo che mostra il tipo di dato di ogni colonna del dataframe
```

```python
nanocomposites_properties.hist(bins=60, figsize=(20,10)) #creo degli istogrammi per tutte le colonne numeriche del dataframe
#y: quante volte quei valori si ripetono
#x: valori delle proprietà (raggruppati nei bin, ho creato 60 bin pen non avere istogrammi troppo frastagliati)
```

```python
#faccio un ciclo for per ogni colonna per vedere il valore più frequente
for colonna in nanocomposites_properties.columns:  #ciclo su tutte le colonne
    print("Colonna: " + colonna)
    counts = nanocomposites_properties[colonna].value_counts()  #value_counts(): metodo che conta quante volte compare ogni valore valore
    #print(counts)
    print("Valore più frequente: " + str(counts.idxmax()) )  #stampo il più comune
```
