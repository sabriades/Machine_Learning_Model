```python
#possibili outliers nel dataframe nanocomposites_properties per ogni colonna
#nanocomposites_properties.plot(kind='box', vert=False)

for colonna in nanocomposites_properties:
  if colonna not in ['Filler', 'Polymer Matrix']:
    nanocomposites_properties[[colonna]].plot(kind='box')


#nanocomposites_properties['Tensile Strain (%)'].plot(kind='box', vert=False)
```

```python
outliers_dict={}
for colonna in nanocomposites_properties:
  if colonna not in ['Polymer Matrix', 'Filler']:
    Q1=nanocomposites_properties[colonna].quantile(0.25)
    Q3=nanocomposites_properties[colonna].quantile(0.75)
    IQR=Q3-Q1
    limite_basso=Q1- 1.5*IQR
    limite_alto=Q3+1.5*IQR
    outliers_colonna=nanocomposites_properties[(nanocomposites_properties[colonna]<limite_basso) | (nanocomposites_properties[colonna]>limite_alto)]
    outliers_dict[colonna] = outliers_colonna
```

```python
for colonna in nanocomposites_properties:
  if colonna not in ['Polymer Matrix', 'Filler', 'Tensile Strength (MPa)']:
    print("outliers di: "+colonna)
    display(outliers_dict[colonna])
```
